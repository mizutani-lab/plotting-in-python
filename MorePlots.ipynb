## Line charts

```python
y = [0, 1, 1, 2, 5, 8]
x = [0, 1, 2, 3, 4, 5]
plt.plot(x, y)
plt.plot([1, 3, 5], [2, 5, 6])
plt.plot([2, 5, 6])
plt.xlabel('x values')
plt.ylabel('Fibonacci')
plt.title('Our first plot')
plt.show()
```

```python
x = np.linspace(0, 10, 100)
```

```python
fig = plt.figure(figsize=(8,6))
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')

# Save figure as file.
fig.savefig('my_figure.png')
```

## Subplots

```python
plt.figure() # optional, but good practice
x = np.linspace(0, 10, 100)

plt.subplot(2, 1, 1) #rows, columns, panel #
plt.plot(x, np.sin(x), '-')

plt.subplot(2, 1, 2) #rows, columns, panel #
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
plt.axis('equal') #create equal aspect ratios
```

## Scatter plot

```python
plt.figure()

#create a random number generator
rng = np.random.RandomState(0) #0 is the seed, generates a random number in [0,1]
for marker in ['o', '.', 'x', '+', '>', '<']:
    plt.plot(rng.rand(5), rng.rand(5), marker)
```

## Histogram

```python
plt.figure()
#samples from normal distribution
data = np.random.randn(1000)
plt.hist(data, bins=20)
```

## Different styles

```python
x = np.arange(-2, 10, .1)
y = .1*x ** 3-x ** 2 + 3 *x +2
fig = plt.figure(dpi=100, figsize=(10,20), tight_layout=True)
available = ['default'] + plt.style.available # list all styles
for i, style in enumerate(available):
    with plt.style.context(style):
        ax = fig.add_subplot(10,3,i+1)
        ax.plot(x,y)
        ax.plot(x,y+2)
        ax.plot(x,y-2)
    ax.set_title(style)
```

## Bar chart

```python
plt.figure()
plt.bar(['a', 'b', 'c'], [10, 50, 20], color=['red', 'blue', 'yellow'])
plt.show()
```

## Box plot

```python
np.random.seed(10)

data_1 = np.random.normal(100, 10, 200)
data_2 = np.random.normal(90, 20, 200)
data_3 = np.random.normal(80, 30, 200)
data_4 = np.random.normal(70, 40, 200)
data = [data_1, data_2, data_3, data_4]

fig = plt.figure() #figsize =(10, 7))

# Creating plot
plt.boxplot(data)

# show plot
plt.show()
```

## Violin plot

```python
plt.figure()
plt.violinplot(data)
plt.show()
```
