## My first plot

```python
import matplotlib.pyplot as plt
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 1, 2, 3, 5]

plt.figure(figsize=(8,8))
plt.plot(x, y)
plt.plot([1,3,5], [4,2,3])

plt.title('My first plot')
plt.xlabel('n')
plt.ylabel('Fibonacci')
plt.show()
```

## Use of numpy

```python
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
```

```python
plt.figure()
plt.plot(x, x)
plt.plot(x, x + 1)
plt.plot(x, x * 2)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

## Visualize Fibonacci Numbers

```python
y = [0, 1]
```

```python
for i in range(100):
    y.append(y[-2] + y[-1])
    #y += [y[-2] + y[-1]]
```

```python
from math import log
```

```python
[log(yy) for yy in y[1:10]]
```

```python
plt.figure()
plt.plot(y[:100])
plt.xscale('log')
plt.yscale('log')
plt.plot([1.6**i for i in range(100)])  # Looks like Fibonacci!
plt.show()
```
